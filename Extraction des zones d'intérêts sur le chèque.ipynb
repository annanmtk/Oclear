Dans ce notebook nous présentons une méthode manuelle d'extration des zones d'intêrets d'un chèque tels que: le montant en chiffre, le montant en lettre, la signature du client, le lieu et la date de la transaction. Il est à noter que comme la positon des chèques ou la tallle peut être différent cette méthode s'adapte plus ou moins aux différents chèques.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from PIL import Image
 
# import the image of check
im = Image.open(r"C:/Users/AIMSSN/Desktop/database/cheque/cheques/cheque65.png")
```

```python
plt.imshow(im) # plotting
```

```python
ima=cv2.cvtColor(np.array(im),cv2.COLOR_RGB2BGR) # tranform image to array
```

```python
montant_chiffre=ima[40:80,525:] # montant en chiffre
plt.figure(figsize=(6, 6))
plt.imshow(montant_chiffre)
```

```python
montant_lettre_1=ima[55:105,120:] # montant en lettre écrit sur la première ligne
plt.figure(figsize=(6, 6))
plt.imshow(montant_lettre_1)
```

```python
montant_lettre_2=ima[95:120,40:] #  montant en lettre écrit sur la deuxième ligne si la première est totalement remplie
plt.figure(figsize=(6, 6))
plt.imshow(montant_lettre_2)
```

```python
lieu=ima[150:175,320:450] # lieu de la transaction 
plt.figure(figsize=(6, 6))
plt.imshow(lieu)
```

```python
date=ima[145:175,490:] # Date de la transaction
plt.figure(figsize=(6, 6))
plt.imshow(date)
```

```python
signature=ima[175:270,425:] # Signature du client
plt.figure(figsize=(6, 6))
plt.imshow(signature)
```
