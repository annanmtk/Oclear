```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('C:/Users/AIMSSN/Desktop/database/cheque/cheques/cheque34.png')
img.shape
plt.imshow(img,'gray')
```

```python
A=[119,86]
B=[663,76]
C=[119,110]
D=[663,99]
w,h=1800,70
```

```python
pts1=np.float32([A,C,B,D])
pts2=np.float32([[0,0],[0,h],[w,0],[w,h]])
```

```python
matrix=cv2.getPerspectiveTransform(pts1,pts2)
out=cv2.warpPerspective(img,matrix,[w,h])
```

```python
plt.imshow(out,'gray')
```

```python
cv2.imwrite('C:/Users/AIMSSN/Desktop/cursive/Cursive_handwriting_recognition-master/out20.png',out)
```
