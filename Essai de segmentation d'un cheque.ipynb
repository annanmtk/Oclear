```python
import pandas as pd
import re
import cv2
import pytesseract
from pytesseract import Output
import numpy as np
from PIL import Image 
from pytesseract import pytesseract
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from IPython import display
```

```python
imgname = "C:/Users/AIMSSN/Desktop/database/cheque/cheques/cheque2.png"
img = cv2.imread(imgname)
mser = cv2.MSER_create()

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
## Do mser detection, get the coodinates and bboxes
coordinates, bboxes = mser.detectRegions(gray)
vis = img.copy()
coords = []
bbs=[]
for coord in coordinates:
    bbox = cv2.boundingRect(coord)
    x,y,w,h = bbox
    #if w< 10 or h < 10 or w/h > 5 or h/w > 5:
    if w< 56 or h < 10 or w/h > 10 or h/w > 1 :
        continue
    coords.append(coord)
    bbs.append((x,y,w,h))
fig, ax = plt.subplots(1, figsize=(18, 16))
ax.imshow(img, cmap='Greys_r')
for (x, y, w, h) in bbs :
    rect = patches.Rectangle((x, y), w, h, fill=False, color="r")
    ax.add_patch(rect)
display.display(plt.gcf())
display.clear_output(wait=True)
```
