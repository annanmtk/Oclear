```python
import imagehash
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import os

def similarity(path1,path2):
    i=0
    l=[path1,path2]
    for i in range(len(l)):        
        ax = plt.subplot(1, 2, i+1)
        plt.imshow(Image.open(l[i]),"gray")
        plt.title("Sign{}".format(i), fontdict={'fontsize': 13, 'color': 'black'})
        plt.axis('off')
    plt.tight_layout()
    plt.show()
    hash0 = imagehash.average_hash(Image.open(path1)) 
    hash1 = imagehash.average_hash(Image.open(path2)) 
    diff = abs(hash0 - hash1) # diff limite est 16 bits experimentalement
    prob=1-(float(diff)/64) # Seuil de prob
    if prob >0.76:
        print("Probability: {}".format(prob).center(50," "))
        print('Authenticated signature'.center(50,"_"))
    else:
        print("Probability: {}".format(prob).center(50," "))
        print('Falsified signature'.center(50,"_"))

```

```python
similarity("1.png","2.png")
```
